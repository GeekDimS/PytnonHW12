```python
import os
import pandas as pd
import matplotlib.pyplot as plt
pd_df = pd.read_csv('company_sales_data.csv')
pd_df.head(12)
```

```python
month_list = pd_df['month_number']
profit_list = pd_df['total_profit']
facecream = pd_df['facecream']
facewash = pd_df['facewash']
toothpaste = pd_df['toothpaste']
bathingsoap = pd_df['bathingsoap']
shampoo = pd_df['shampoo']
moisturizer = pd_df['moisturizer']

month_list
```

```python
plt.xticks(month_list)
plt.plot(facecream, label = 'Продажи крема для лица')
plt.plot(facewash, label = 'Продажи мыла для лица')
plt.plot(toothpaste, label = 'Продажи зубной пасты')#plt.plot(bathingsoap, label = 'Продажи крема для лица')
plt.plot(bathingsoap, label = 'Продажи мыла для купания')
plt.plot(shampoo, label = 'Продажи шампуня')
plt.plot(moisturizer,label = 'Продажи увлажнителя')
plt.title('Данные о продажах продуктов:')
plt.ylabel('Число продаж')
plt.xlabel('Месяцы')
plt.grid(linestyle = '--')
plt.legend(bbox_to_anchor=[1, 1], loc = 'best')#Располоагаем легенду справа от графика
plt.show
```

3. Рассчитайте общие данные о продажах за год для каждого продукта и покажите их с помощью круговой диаграммы

```python
plt.pie((sum(facecream), sum(facewash),sum(toothpaste),sum(bathingsoap),sum(shampoo),sum(moisturizer)),
       labels = ('Крем для лица','Мыло для лица', 'Зубная паста', 'Мыло для купания', 'Шампунь', 'Увлажнитель'),
       autopct = '%.1f', labeldistance = 1.1, radius = 2, rotatelabels = False, )
plt.title('Общие данные о продажах за год для каждого продукта',pad =100)
plt.legend(bbox_to_anchor=[2.5, 1], loc = 'best')#Располоагаем легенду справа от графика
```
